```python
%matplotlib inline
import ipdb
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.svm import SVC
```

```python
iris = load_iris()
y = iris.target
pca = PCA(n_components=2)
X = pca.fit_transform(iris.data)

svm = SVC(gamma=0.7,kernel="linear")
svm.fit(X,y)

h=0.2
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)

cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
cmap_grey = ListedColormap(["#808080"])
fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_size_inches(10,10)
fig.tight_layout()
ax1.contourf(xx, yy, np.ones_like(Z), cmap=cmap_grey)
ax1.scatter(X[:,0], X[:,1], c=y, cmap=cmap_bold)
ax1.axis("off")
ax1.set_title("(A) Three classes")
x0,x1 = ax1.get_xlim()
y0,y1 = ax1.get_ylim()
ax1.set_aspect((x1-x0)/(y1-y0))
ax2.contourf(xx, yy, Z, cmap=cmap_light)
# Plot also the training points
ax2.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
ax2.axis('off')
ax2.set_title("(B) Support vector machine with linear kernel")
x0,x1 = ax2.get_xlim()
y0,y1 = ax2.get_ylim()
ax2.set_aspect((x1-x0)/(y1-y0))
fig.savefig(
    "../../Programming/MasterProject/thesis/figs/supervised_PCA.png",
    format="png",
    transparent=True,
    bbox_inches="tight")
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
np.random.seed(0)

def f(x):
    return np.sin(2 * np.pi * x)
x_plot = np.linspace(0, 1, 100)
n_samples = 100
degree=3
label='cubic regression'
X = np.random.uniform(0, 1, size=n_samples)[:, np.newaxis]
y = f(X) + np.random.normal(scale=0.3, size=n_samples)[:, np.newaxis]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)

est = make_pipeline(PolynomialFeatures(degree), LinearRegression())
est.fit(X_train, y_train)


fig,(ax1,ax2) = plt.subplots(1,2)
fig.set_size_inches(10,10)
fig.tight_layout()

#ax1 = plt.gca()
ax1.scatter(X_train, y_train, s=10)
ax1.set_ylim((-2, 2))
ax1.set_xlim((0, 1))
ax1.set_ylabel('y')
ax1.set_xlabel('x')
ax1.set_title("(A) Random sine with noise")
x0,x1 = ax1.get_xlim()
y0,y1 = ax1.get_ylim()
ax1.set_aspect((x1-x0)/(y1-y0))

ax2.plot(x_plot, f(x_plot), color='green', label="true curve")
ax2.scatter(X_train, y_train, s=10)
ax2.plot(x_plot, est.predict(x_plot[:, np.newaxis]), color='red', label=label)
ax2.set_ylim((-2, 2))
ax2.set_xlim((0, 1))
ax2.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_title("(B) cubic regression")
ax2.legend(loc='upper right')  #, fontsize='small')
x0,x1 = ax2.get_xlim()
y0,y1 = ax2.get_ylim()
ax2.set_aspect((x1-x0)/(y1-y0))
fig.savefig(
    "../../Programming/MasterProject/thesis/figs/polinomial_regression.png",
    format="png",
    transparent=True,
    bbox_inches="tight")
```
